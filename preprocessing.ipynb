```python
#Load dan Merge Data
import pandas as pd

# Load data
df_fingrid = pd.read_csv("fileExtracted/extractionDatasetFinGrid.csv", delimiter=";")
df_fmi = pd.read_csv("fileExtracted/finland_country_level_weather.csv", delimiter=";")
```

```python
print(df_fingrid.columns.tolist())
print(df_fmi.columns.tolist())
```

```python
print(df_fingrid.columns)
print(df_fmi.columns)  # See what columns are available
```

```python
# Konversi timestamp ke datetime
df_fingrid["timestamp"] = pd.to_datetime(df_fingrid["endTime"]).dt.tz_localize(None)
df_fmi["timestamp"] = pd.to_datetime(df_fmi["Time (UTC)"])
```

```python
print(df_fmi.timestamp)
print(df_fingrid.timestamp)
```

```python
# Merge berdasarkan timestamp
df_merged = pd.merge(df_fingrid, df_fmi, on="timestamp", how="inner")
df_merged.head()
```

```python
print(df_merged.columns)
print(df_merged.value)
```

```python
#Hapus duplikasi
df_merged.drop_duplicates(subset=["timestamp"], keep="first", inplace=True)

#Handling Missing Values
# Cek missing values
print(df_merged.isnull().sum())

# Imputasi (misal: forward fill untuk time series)
df_merged.ffill(inplace=True)
```

```python
#Feature Engineering
# Time-based features
df_merged["hour"] = df_merged["timestamp"].dt.hour
df_merged["day_of_week"] = df_merged["timestamp"].dt.dayofweek
df_merged["month"] = df_merged["timestamp"].dt.month
df_merged["is_weekend"] = df_merged["day_of_week"].isin([5,6]).astype(int)

# Lagged Features (untuk time series)
df_merged["lag_1h_consumption"] = df_merged["value"].shift(1)
df_merged["rolling_mean_24h"] = df_merged["value"].rolling(window=24).mean()

# Interaksi Fitur (misal: suhu x waktu)
df_merged["temp_hour_interaction"] = df_merged["Air temperature"] * df_merged["hour"]
electricity_price_per_MW = 40  # Harga dalam euro per MW 2025
df_merged["price"] = df_merged["value"] * electricity_price_per_MW
```

```python

# Debugging: Cek nama kolom
print("Available columns in df_merged:", df_merged.columns.tolist())
```

```python
%pip install scikit-learn
```

```python
#Normalisasi
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numeric_cols = ["Air temperature", "Wind speed", "value", "price"]
df_merged[numeric_cols] = scaler.fit_transform(df_merged[numeric_cols])
```

```python
#Split Data untuk Forecasting
train_size = int(len(df_merged) * 0.7)
val_size = int(len(df_merged) * 0.15)
test_size = len(df_merged) - train_size - val_size

train_df = df_merged.iloc[:train_size]
val_df = df_merged.iloc[train_size:train_size+val_size]
test_df = df_merged.iloc[train_size+val_size:]
```

```python
# 1. Reset index
train_df = train_df.reset_index(drop=True)
val_df = val_df.reset_index(drop=True)
test_df = test_df.reset_index(drop=True)

# 3. Buat DataFrame baru (opsional)
train_df = pd.DataFrame(train_df.values, columns=train_df.columns)
val_df = pd.DataFrame(val_df.values, columns=val_df.columns)
test_df = pd.DataFrame(test_df.values, columns=test_df.columns)

# 2. Perbaiki nama kolom
train_df.columns = [str(col) for col in train_df.columns]
val_df.columns = [str(col) for col in val_df.columns]
test_df.columns = [str(col) for col in test_df.columns]
```

```python
print(train_df.dtypes)  # Pastikan semua kolom memiliki tipe data valid
print(train_df.columns) # Pastikan nama kolom adalah string

# Hapus kolom yang tidak valid (contoh: kolom bernama 'level_0' atau 'index')
invalid_cols = [col for col in train_df.columns if not isinstance(col, str)]
train_df.drop(columns=invalid_cols, inplace=True)
```

```python
print(train_df.columns)
print(train_df.dtypes)
print(train_df.head().to_dict())
```

```python
train_df = train_df.copy()
val_df = val_df.copy()
test_df = test_df.copy()
```

```python
# Simpan sebagai CSV
train_df.to_csv("processed_data/train.csv", encoding="utf-8")
val_df.to_csv("processed_data/val.csv", encoding="utf-8")
test_df.to_csv("processed_data/test.csv", encoding="utf-8")
```
